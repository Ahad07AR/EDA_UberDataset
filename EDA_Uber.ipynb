#**Data Cleaning**
Steps:
1. Dropping Missing Values
2. Rename column headers
3. Splitting the Start_date and End_start into Hour, Day of the week, month and Weekend.


```python
# Building a copy of the dataset as 'uber'
uber = uber_trips.copy()
uber.head()
```

```python
uber.isnull().values.any()
```

```python
uber.isnull().values.sum()
```

#Dropping Null values
(Note- 'inplace=True' means that the changes will be made directly to the uber DataFrame itself, rather than creating a new DataFrame with the rows removed.)

```python
uber.dropna(inplace=True)
uber
```

Renaming Columns

```python
uber.rename(columns = {'START_DATE*':'Start_Date', 'END_DATE*':'End_Date',
                              'CATEGORY*':'Category', 'START*':'Start', 'STOP*':'Stop', 'MILES*':'Miles', 'PURPOSE*':'Purpose'}, inplace = True)
```

```python
uber.head(10)
```

```python
# Importing datetime and Calender
import datetime
import calendar
```

```python
#Split the start_date and end_date into hour,day,day of the week, month and weekday
uber['Start_Date'] = pd.to_datetime(uber['Start_Date'], format="%m/%d/%Y %H:%M")
uber['End_Date'] = pd.to_datetime(uber['End_Date'], format="%m/%d/%Y %H:%M")
hour=[]
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in uber['Start_Date']:
    hour.append(x.hour)
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
uber['HOUR']=hour
uber['DAY']=day
uber['DAY_OF_WEEK']=dayofweek
uber['MONTH']=month
uber['WEEKDAY']=weekday
```

```python
uber.head()
```

- The table above displays splitting of the Start_date and End_start into Hour, Day, Day of the week, month and weekday

#Q.1 Most popular start destination for the uber drivers?

```python
#'uber' is the DataFrame and 'Start' column is the starting points

# Step 1: Extract the 'Start' column and count the occurrences of each start location
start_counts = uber['Start'].value_counts()

# Step 2: Identify the most popular start location
most_popular_start = start_counts.idxmax()
most_popular_count = start_counts.max()

# Display the result
print(f"The most popular start destination is: {most_popular_start} with {most_popular_count} trips.")
```

Cary city happens to be the most popular start destination

# Q.2 What is the most popular stop/drop off destination for the uber drivers?

```python
# Same as above but with Stop column

stop_counts = uber['Stop'].value_counts()

most_popular_stop = stop_counts.idxmax()
most_popular_count = stop_counts.max()

print(f"The most popular start destination is: {most_popular_stop} with {most_popular_count} trips.")
```

Cary is also the most popular drop off location

# Q.3 Most frequent route taken by the uber drivers

```python

# Combine the 'Start' and 'Stop' columns to create a 'Route' column
uber['Route'] = uber['Start'] + " to " + uber['Stop']

# Count the occurrences of each route
route_counts = uber['Route'].value_counts()

# Identify the most frequent route
most_frequent_route = route_counts.idxmax()
most_frequent_count = route_counts.max()

print(f"The most frequent route is: {most_frequent_route} with {most_frequent_count} trips.")
```

The most frequent route traveeled is from Cary to Morrisville

```python
#function to display values on each plot
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center")
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
#plot for category column
a = sns.countplot(x='Category', data=uber, palette=['#432371', 'gold'])
plt.title('Category of Trips', weight='bold').set_fontsize('16')
show_values(a)
```

There are two categories into which Uber usage may be divided: business and personal.
It is clear from the chart that the majority of users use Uber for business.

#Q. Weekend with highest no. of trips.

```python

order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday","Sunday"] # Define 'order' before it is used
g=sns.catplot(data=uber, x="WEEKDAY", kind="count", palette=sns.color_palette(['#432371']), aspect=15/8,order=order)
plt.title('Weekday With Highest Trip', weight='bold').set_fontsize('16')
#function to display values
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.025,
            p.get_height() * 1.02,
            '{0:.2f}'.format(p.get_height()),
            color='black', rotation='horizontal', size='large')
```

Friday is the day which have the highest number of trips every week.
